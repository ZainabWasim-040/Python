```python
months=[("January",31),("February",28),("March",31),("April",30),("May",31),("June",30),("July",31),("August",31),
          ("September",30),("October",31),("November",30),("December",31)]
start_day=2
for name, days in months:
    print("\n", name)
    print("M\tT\tW\tT\tF\tS\tS")
    print("\t" * start_day, end="")
    for d in range(1, days + 1):
        print(d, end="\t")
        if (d + start_day) % 7 == 0:
            print()
    start_day = (start_day + days) % 7
```
